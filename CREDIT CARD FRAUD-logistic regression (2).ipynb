```python
#splitting data into independent & dependent variables
x=data.drop(['Class'],axis=1)
x.head()
x.shape

y=data['Class']
y.head()
y.shape
```

```python
#undersampling
from imblearn import under_sampling
from imblearn.under_sampling import NearMiss
```

# implimenting undersampling to handle imbalanced dataset

```python
nm=NearMiss()
x_i,y_i=nm.fit_sample(x,y)
x_i.shape
y_i.shape
```

# logistic reg

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_i,y_i,test_size=.20,random_state=40)
print("x_train :",x_train.shape)
print("x_test :",x_test.shape)
print("y_train :",y_train.shape)
print("y_test :",y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
y_predicted=logmodel.predict(x_test)
y_predicted
```

```python
logmodel.score(x_test,y_test)
```

```python
logmodel.coef_
```

```python
logmodel.intercept_
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_predicted)
print("confusion_matrix: ")
print(cm)
print("accuracy_score:",accuracy_score(y_test,y_predicted))
print(classification_report(y_test,y_predicted))
```
